## Proving Sum of Weights = 1 in the Incremental Sum Rule

We call this a weighted average because the sum of the weights $$(1 - \alpha)^n + \sum_{i=1}^{n} \alpha(1-\alpha)^{n-i} = 1$$

First we extract the constant out of the series
$$(1 - \alpha)^n + \alpha(1-\alpha)^{n}\sum_{i=1}^{n} (1-\alpha)^{-i}$$ 

Then we rewrite, and find that we have a finite geometric series

$$(1 - \alpha)^n + \alpha(1-\alpha)^{n}\sum_{i=1}^{n} (\frac{1}{1-\alpha})^{i} $$

Using the formula we can find the sum of the finite geometric series 

$$Sum = \frac{\text{first term} * (1 - rate^n)}{1-rate} $$

Using the formula we can simplify the sum

$$(1 - \alpha)^n + \alpha(1-\alpha)^{n}\frac{1}{1-\alpha}\frac{1-(\frac{1}{1-\alpha})^n}{1-(\frac{1}{1-\alpha})}$$

Then we multiply and simplify

$$(1 - \alpha)^n + \frac{\alpha(1-\alpha)^{n} - \alpha}{-\alpha}$$

Then we divide and simplify

$$(1 - \alpha)^n - (1-\alpha)^n + 1$$

Finally we arrive with answer of 1 proving that the sum of the weights in the incremental sum run is equal to 1

$$ 1 $$

## Exercise 2.8

In this 10 armed bandit problem that uses UCB the first 10 steps are used to try each of the 10 actions. On the 11 action the all values of the $c\sqrt{\frac{ln(t)}{N_t(a)}}$ will be the same so UCB will select the action that provided the highest reward. When averaged over many runs the action with the highest action value will have a high expected value. The reason that there is a decrease in reward on the next step is that the $c\sqrt{\frac{ln(t)}{N_t(a)}}$ is higher for all actions not performed on the 11 time step. So most of the time, depending on how high the estimated value of the 11th action is, it will perform an action other than the action performed on the 11th time step which has a high probability of being lower causing the reward to drop.
